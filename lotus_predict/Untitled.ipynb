```python
news_correct = 0
entertain_correct = 0
count = 0 
data_check ={}
total_news = 0
total_entertain = 0

name_video = []
links = []
labels = []
predicts = []


for index in range(total-1):
    print(model_predict['VideoId'][index][31:])
    check_name = data.label[data.name[data.name == model_predict['VideoId'][index][31:]].index[0]]
    
    if check_name == ' entertainment' or check_name ==' not_entertainment' :
        
        if model_predict['VideoId'][index][31:] in data['name'].values:

            print('Model predict name ', model_predict['VideoId'][index][31:])
            print('Predict model ', model_predict['LabelConfidencePairs'][index])

            label_of_hand = data.label[data.name[data.name == model_predict['VideoId'][index][31:]].index[0]].strip()
            name_in_hand = data.name[data.name == model_predict['VideoId'][index][31:]].values[0]



            print('Name of hand_labels ', name_in_hand)
            print('Data of hand_labels ', label_of_hand)
            if label_of_hand =='entertainment':
                total_entertain+=1
            if label_of_hand == 'not_entertainment':
                total_news+=1
            count +=1
            if model_predict['LabelConfidencePairs'][index].lower() == label_of_hand:
                entertain_correct +=1
            elif model_predict['LabelConfidencePairs'][index].lower() == 'news program' and label_of_hand == 'not_entertainment':
                news_correct+=1
            else:
                predicts.append(' '+model_predict['LabelConfidencePairs'][index].lower())
                labels.append(' '+label_of_hand)
                name_video.append(name_in_hand)
                links.append(data.link[data.name == model_predict['VideoId'][index][31:]].values[0])
            print('...............................')
```
